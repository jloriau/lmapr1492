```python
from pymatgen.ext.matproj import MPRester 
from pymatgen.electronic_structure.bandstructure import BandStructure
from pymatgen.electronic_structure.plotter import BSPlotter
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
```

```python
with MPRester("pte9sXqn7RBHRMrR72d") as m:

    # Bandstructure for material id
    bandstructure = m.get_bandstructure_by_material_id("mp-1018091")
    

```

```python
#bande interdite
bandgap = bandstructure.get_band_gap()

if(bandgap.get('direct')==False):
    print("La bande interdite n'est pas directe")

else:
    print("La bande interdite est directe ")


print("Energie de la bande interdite :", bandgap.get('energy'), "eV")
```

```python
bandplot = BSPlotter(bandstructure)
bandplot.get_plot(vbm_cbm_marker=True)
```

```python
#dispersions pour la dernire bande de valence 

#difference d'energie entre chaque branche pour la derniere bande de valence 
e0 = energie.get('1')[0][18]
energie0 = e0[0]-e0[-1]
print("difference d'energie entre -M: ", energie0)

e1 = energie.get('1')[1][18]
energie1 = e1[0]-e1[-1]
print("difference d'energie entre M-K: ", energie1)

e2 = energie.get('1')[2][18]
energie2 = e2[0]-e2[-1]
print("difference d'energie entre K-: ", energie2)

e3 = energie.get('1')[3][18]
energie3 = e3[0]-e3[-1]
print("difference d'energie entre -A: ", energie3)

e4 = energie.get('1')[4][18]
energie4 = e4[0]-e4[-1]
print("difference d'energie entre A-L: ", energie4)

e5 = energie.get('1')[5][18]
energie5 = e5[0]-e5[-1]
print("difference d'energie entre L-H: ", energie5)

e6 = energie.get('1')[6][18]
energie6 = e6[0]-e6[-1]
print("difference d'energie entre H-A|L: ", energie6)

e7 = energie.get('1')[7][18]
energie7 = e7[0]-e7[-1]
print("difference d'energie entre A|L-M|K: ", energie7)

e8 = energie.get('1')[8][18]
energie8 = e8[0]-e8[-1]
print("difference d'energie entre M|K-H: ", energie8)

#distance entre chaque branche d'energie 

dist0 = distance[1]-distance[0]
dist1 = distance[3]-distance[1]
dist2 = distance[5]-distance[3]
dist3 = distance[7]-distance[5]
dist4 = distance[9]-distance[7]
dist5 = distance[11]-distance[9]
dist6 = distance[13]-distance[11]
dist7 = distance[14]-distance[13]
dist8 = distance[15]-distance[14]

#pente entre chaque branche pour la derniere bande de valence

pente0 = energie0/dist0
pente1 = energie1/dist1
pente2 = energie2/dist2
pente3 = energie3/dist3
pente4 = energie4/dist4
pente5 = energie5/dist5
pente6 = energie6/dist6
pente7 = energie7/dist7
pente8 = energie8/dist8

pentes = np.array([pente0, pente1,pente2,pente3,pente4,pente5,pente6,pente7,pente8])
print(pentes )

penteabs = [abs(pente0), abs(pente1),abs(pente2),abs(pente3),abs(pente4),abs(pente5),abs(pente6),abs(pente7),abs(pente8)]

print('La dispersion maximale pour la dernire bande de valence est en A-L : ',  max(penteabs) )
print('La dispersion minimale pour la dernire bande de valence est en K- : ',  min(penteabs) )
```

Nous allons dterminer la masse effective au sommet de la dernire bande de valence et  la base de la premire bande de conduction. Le tenseur des masses effectives est symtrique, donc on peut le rduire  3 composantes indpendantes sur la diagonale. En supposant une dispersion parabolique, la masse effective est isotrope et donc le tenseur des masses effectives peut finalement se rduire  un scalaire.

On peut calculer la masse effective comme :$ m_* =  \frac{\hbar^2}{\frac{d^2\epsilon}{dk^2}}$

Avec $\hbar$, la constante de Planck, $\epsilon$, l'nergie et k, le vecteur d'onde.

On calcule la drive seconde d'une parabole : $ ax^2+bx+c$. La drive donne donc 2a.

Pour avoir une masse en kg, il faut diviser par un eV et le carr d'un $$.
Pour une masse en masse d'lectron, il faut diviser par la masse de l'lectron dans le vide.

```python
#Constante de Planck
hbar = 6.58212e-16 


# Masse effective pour le maximum de la bande de valence

vbm =  bandstructure.get_vbm()
ev = vbm['energy']
kv = vbm['kpoint'].cart_coords

mv = (hbar**2)/(2*ev)*(kv[0]**2+kv[1]**2+kv[2]**2)

print("La masse effective au sommet de la bande de valence  en supposant une dispersion parabolique vaut :", mv, '[kg]')

# Masse effective pour le minimum de la bande de conduction
cbm =  bandstructure.get_cbm()
ec = cbm['energy']
kc = cbm['kpoint'].cart_coords

mc = (hbar**2)/(2*ec)*(kc[0]**2+kc[1]**2+kc[2]**2)

print("La masse effective  la base de la bande de conduction  en supposant une dispersion parabolique vaut :", mc, '[kg]')
```
