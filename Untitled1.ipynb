```python
#import
from pymatgen.ext.matproj import MPRester 
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("pte9sXqn7RBHRMrR72d") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-1018091")
    
    w = CifWriter(structure);
    w.write_file('mystructuremp-1018091.cif')
```

```python
# objet XRDCalculator pour un rayonnement X incident Cu Ka de longueur d’onde λ=1.54060 Å 

xrd = XRDCalculator(1.54060)  

# Modèle de diffraction

modele = xrd.get_pattern(structure)
```

```python
#pic 1 de diffraction

#hkl du pic 1

hkl1 = modele.hkls[0]

#angle du pic 1

angle1 = modele.x[0]

#intensite du pic 1

intensity1 = modele.y[0]

print("Premier pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl1)
print("Angle : 2*theta =", angle1,"°")
print("Intensité :", intensity1)
```

```python
#pic 2 de diffraction

#hkl du pic 2

hkl2 = modele.hkls[1]

#angle du pic 2

angle2 = modele.x[1]

#intensite du pic 2

intensity2 = modele.y[1]

print("Premier pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl2)
print("Angle : 2*theta =", angle2,"°")
print("Intensité :", intensity2)
```

```python
#pic 3 de diffraction

#hkl du pic 3

hkl3 = modele.hkls[2]

#angle du pic 3

angle3 = modele.x[2]

#intensite du pic 3

intensity3 = modele.y[2]

print("Premier pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl3)
print("Angle : 2*theta =", angle3,"°")
print("Intensité :", intensity3)
```
